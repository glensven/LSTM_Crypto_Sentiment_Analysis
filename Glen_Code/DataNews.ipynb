```python
import requests
import pandas as pd
import json
from path import Path
from dotenv import load_dotenv
import os
import requests
import datanews
```

```python
# Laod .env environment
load_dotenv()
```

```python
# Load API Key
datanews.api_key = os.getenv("datanews.api_key")
type(datanews.api_key)
```

```python
response = datanews.news(q='Bitcoin', language=['en'], from_date = '2020-01-01', to_date = '2020-02-01')
articles = response['hits']
articles
```

```python
help(datanews.news)
```

```python
# Transform the response dictionary to a DataFrame
bitcoin_df = pd.DataFrame(articles)

bitcoin_df.head(20)
```

```python
# Grab Published Date, Title, and Content in New Dataframe 
bitcoin_summary_df = bitcoin_df[['pubDate','title','content']]
bitcoin_summary_df.pubDate = pd.to_datetime(bitcoin_summary_df.pubDate)
bitcoin_summary_df.set_index('pubDate', inplace = True)
bitcoin_summary_df.columns = ['Title','Content']
bitcoin_summary_df.index.names = ['Published Date']
bitcoin_summary_df
```
