```python
data_df = pd.DataFrame(data= [['','','','','','']], columns = ['id', 'content', 'title', 'media-type', 'source', 'published'])
data_df
```

```python
# Load json
data = Path('C:\\Users\\Glen\\Documents\\GitHub\\NLP-Crypto-Project\\Glen_Code\\test.jsonl')


with open(data, 'rb') as f: # opening file in binary(rb) mode    
       for item in json_lines.reader(f):
            list_keys = []
            list_values = []
            for k, v in item.items():
                list_keys.append(k)
                list_values.append(v)
            if "Bitcoin" in item['content']:
                new_df = pd.DataFrame([list_values], columns = list_keys)
                data_df = pd.concat([data_df, new_df], ignore_index = True)
                #data_df.append(item, ignore_index = True) 
```

```python
data_df = data_df.dropna()
data_df['published'] = pd.to_datetime(data_df['published'])
data_df = data_df.dropna()
data_df = data_df.set_index('published')
data_df
```
